```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('total_cases.csv')
print(df)
```

```python
df.isna().sum()
```

```python
df.isna().sum()/len(df)*100
```

```python
df.dropna(axis=0)
```

```python
fig = px.bar(df, x='date', y='Albania',title = "Total Covid cases in ALBANIA")
fig.show()
```

```python
fig = px.bar(df, x='date', y='Afghanistan',title = "Total Covid cases in Afghanistan")
fig.show()
```

```python
fig = px.scatter(df, x='date', y='Afghanistan' ,title = "Total Covid cases in Afghanistan")
fig.show()
```

```python
fig = px.line(df, x='date', y=['Afghanistan','Albania','China'] ,title = "Total Covid cases")
fig.show()
```

```python
fig = px.bar(df, x='date', y=['Afghanistan','Albania','Bahamas'],title = "Total Covid cases in ALBANIA")
fig.show()
```

```python
fig=px.imshow(df.corr(numeric_only=True),zmin=-1,zmax=1 , color_continuous_scale='rdbu')
fig.show()
```
